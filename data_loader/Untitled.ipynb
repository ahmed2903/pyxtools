```python

import numpy as np
import matplotlib.pyplot as plt

a = np.zeros((10,10))
```

```python
a[3:7, 3:7] = 1
```

```python

plt.figure()
plt.imshow(a)
```

```python
coords = np.where(a>0)
coords = np.transpose(coords)
```

```python
coords.shape
```

```python

a[coords[5,0], coords[5,1]] = 10
```
